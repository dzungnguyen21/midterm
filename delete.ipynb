```python
import cv2 
import matplotlib.pyplot as plt


main_image = cv2.imread("Finding/1.jpg")
template = cv2.imread("Template/find/duck.png")

main_gray = cv2.cvtColor(main_image, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

main_edges = cv2.Canny(main_gray, 100, 300)
template_edges = cv2.Canny(template_gray, 50, 100)


plt.figure(figsize=(80, 40))
plt.subplot(2, 1, 1)
plt.imshow(main_edges, cmap='gray')
plt.show()
plt.subplot(2, 1, 2)
plt.imshow(template_edges, cmap='gray')
plt.show()

```
